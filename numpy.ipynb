#THE BASICS :

#initialize array

```python
a=np.array([1,2,3],dtype="int16") # with assigning to dtype i,e keyword argument we  can make it possible to store our data in specified number of bytes
#but it need to fit into that specified byte of storage; 
print("one dimensional array :",a)
```

```python
b=np.array([[1.0,2.0,3.0,5.8],[2.1,3,4.2,4.5]])
print("two dimensional array:",b)
```

```python
#Get Dimension
print("dimension of a :",a.ndim)
print("dimension of b :",b.ndim)
```

```python
#Get Shape
print("shape of a :",a.shape) #its a vector so its tellings us one dimension 
print("shape of b :",b.shape)
```

```python
#Get  Type
print("a type is :",a.dtype)

#Get Size
print("size of a :",a.size) # # of element in the array.i,e length of array;
```

```python
#total size
print("a item size :",a.itemsize) # size of single element in the array .i,e in bytes;
print("total size is a:",a.size*a.itemsize)
```

```python
#Get specific element [row,column] -ve index is possible similar to list;
b[1,2]

#Get specific row [row,:]#we use slicing(:) operation
print("row :",b[1,0:1]);

#Get specific column[:,column]
print("column :",b[:,1])
```
